```python
manuSet = set(ledger['设备供应商'].tolist())
print(len(manuSet))
manuCount = {i:0 for i in manuSet}
for i in range(len(ledger)):
    manuCount[ledger['设备供应商'][i]] += 1

manuCount
```

```python
manu_substation = {i:set() for i in manuSet}
for i in range(len(ledger)):
    manu_substation[ledger['设备供应商'][i]].add(ledger['所在变电站'][i])

manu_substation_count = {i:len(manu_substation[i]) for i in manu_substation}
manu_substation
```

```python
manu_substation_count
```

```python
manuCount['山东现代达驰电工电气股份有限公司']
```

```python
manu_substation_defect_ratio = {}
for i in manu_substation:
    manu_substation_defect_ratio[i] = {}
    for sub in manu_substation[i]:
        manu_substation_defect_ratio[i][sub] = {'all':0, 'bad':0, 'all_equip': set(), 'bad_equip':set() }

manu_substation_defect_ratio
```

```python
for i in range(len(data)):
    manu_substation_defect_ratio[data['设备供应商'][i]][data['所在变电站'][i]]['bad'] += 1
    manu_substation_defect_ratio[data['设备供应商'][i]][data['所在变电站'][i]]['bad_equip'].add(data['设备编号'][i])
```

```python
for i in range(len(ledger)):
    manu_substation_defect_ratio[ledger['设备供应商'][i]][ledger['所在变电站'][i]]['all'] += 1
    manu_substation_defect_ratio[ledger['设备供应商'][i]][ledger['所在变电站'][i]]['all_equip'].add(ledger['设备编号'][i])
```

```python
manu_substation_defect_ratio
```

```python
manu_bad_ratio = {i:0 for i in manu_substation_defect_ratio}
for manu in manu_substation_defect_ratio:
    all_sub = 0
    bad = 0
    for sub in manu_substation_defect_ratio[manu]:
        all_sub += 1
        if len(manu_substation_defect_ratio[manu][sub]['bad_equip']) > 0:
            bad += 1
    manu_bad_ratio[manu] = bad / all_sub

manu_bad_ratio
```
